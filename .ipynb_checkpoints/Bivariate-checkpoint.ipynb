### Effect Size

We often refer to the relationship of a feature with the label as the effect size of that feature.

### Pearson correlation

Is a statistical measure of effect size that indicates how much two numeric variables infuelnce each other. (Feature and Label are both normal distributed)

Good/Strong correlation?

    - Small effect size:  .10 < r < .29
    - Medium effect size: .30 < r < .49
    - Large effect size:  .50 < r

### Assumptions

#### 1.- Continuous data

If the data are not continuous/interval, that will often result in violations of the second two assumptions.

#### 2.- Linear relationship

Means that the relationship between two variables can be represented by a straight line.

#### 3.- Homoskedastic relationship

Refers to the consistency of the relationship acros all values of x and y.

### P-Value

Is the probability that future data collections will have results at least as extreme as those observed in the test data.

We like low p-values because it simply means we can trust the results



So, to assess the statistical significance of your correlation, you need to look at the p-value that is calculated alongside the Pearson coefficient, which can be interpreted as follows:

 If the p-value is low (generally less than 0.05), then your correlation is statistically significant, and you can use the calculated Pearson coefficient.

 If the p-value is not low (generally higher than 0.05), then your correlation is not statistically significant (it might have happened just by chance) and you should not rely upon your Pearson coefficient.

```python
from scipy import stats
r, p = stats.pearsonr(df.charges, df.age)
r, p #Pearson, P-Value
```

It is extremely likely that we'll see that correlation again next time we collect 1300 more samples (because p-values is 0)
